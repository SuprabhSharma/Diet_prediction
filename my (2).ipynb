```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    stratify=y,
    random_state=42
)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(class_weight='balanced', 
    n_estimators=150,        # moderate trees
    max_depth=6,             # limit tree depth strongly
    min_samples_split=20,    # prevent tiny splits
    min_samples_leaf=10,     # avoid leaf overfitting
    max_features=0.6,        # random feature subset
    bootstrap=True,
    random_state=42
)


rf_model.fit(X_train, y_train)



```

```python
print("Train Accuracy:", rf_model.score(X_train, y_train))
print("Test Accuracy:", rf_model.score(X_test, y_test))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_model = AdaBoostClassifier(
    n_estimators=120,      # moderate number
    learning_rate=0.3,     # prevents overfitting
    random_state=42
)

ada_model.fit(X_train, y_train)
```

```python
train_acc = ada_model.score(X_train, y_train)
test_acc = ada_model.score(X_test, y_test)

print("Training Accuracy:", train_acc)
print("Testing Accuracy:", test_acc)
```

```python
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=210,        # slightly fewer trees
    learning_rate=0.035,     
    max_depth=3,             # keep shallow trees
    subsample=0.72,          # add randomness
    colsample_bytree=0.75,
    gamma=2.3,               # stronger split penalty
    reg_alpha=1.2,           # L1 regularization 
    reg_lambda=2.8,          # L2 regularization 
    random_state=42,
    eval_metric="mlogloss"
)

xgb_model.fit(X_train, y_train)
```

```python
train_acc = xgb_model.score(X_train, y_train)
test_acc = xgb_model.score(X_test, y_test)

print("Train Accuracy:", train_acc)
print("Test Accuracy:", test_acc)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Get feature importance
importance = xgb_model.feature_importances_

feature_imp = pd.DataFrame({
    "Feature": X.columns,
    "Importance": importance
}).sort_values("Importance", ascending=False)

print(feature_imp.head(10))
```

```python
plt.figure(figsize=(10,6))
plt.barh(feature_imp["Feature"][:10],
         feature_imp["Importance"][:10])
plt.gca().invert_yaxis()
plt.title("Top Important Features")
plt.show()
```
